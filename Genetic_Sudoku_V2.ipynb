```python
  # Crossover Function
import random
def crossover(pa, ma):
   parents = [ma,pa]
   bro = []
   sis = []
   for i in range(9):
     bro_row = []
     sis_row = []
     for j in range(9):
        x = random.randint(0,1)
        bro_row.append(parents[x][i][j])
        sis_row.append(parents[1-x][i][j])
     bro.append(bro_row)
     sis.append(sis_row)
   return bro, sis
 
import random
def mutation(chr, sudoku, mutation_rate = .1):
    genetic_freak = []
    for i in range(9):
        row = []
        for j in range(9):
            x = random.random()
            if x <= mutation_rate:
                row.append(random.choice(list(sudoku[i][j])))
            else: row.append(chr[i][j])
        genetic_freak.append(row)
    return genetic_freak
    
def reproduce(Population, sudoku, mut_prob = 0.021, new_pop_size = 100, tournament_size = 10):
    new_pop = []
    for i in range(new_pop_size):
        x = random.random()
        row = Population.loc[Population.Cumulative_Probability >= x].nsmallest(1, 'Cumulative_Probability')
        pa = row.values[0][0]
        x = random.random()
        row = Population.loc[Population.Cumulative_Probability >= x].nsmallest(1, 'Cumulative_Probability')
        ma = row.values[0][0]
#         row = Population.loc[random.sample(range(len(Population)),tournament_size) ].nlargest(1, 'fitness_value')
#         pa = row.values[0][0]
#         row = Population.loc[random.sample(range(len(Population)), tournament_size)].nlargest(1, 'fitness_value')
#         ma = row.values[0][0]
        bro, sis = crossover(pa, ma)
        bro = mutation(bro, sudoku, mut_prob)
        sis = mutation(sis, sudoku, mut_prob)
        new_pop.append(bro)
        new_pop.append(sis)
    return new_pop   
```

```python
def transpose(sudoku):
  trans_sudoku = [[9,0,0,0,2,3,0,0,0],
          [1,0,6,0,0,0,9,0,0],
          [0,3,0,8,9,0,0,0,7],
          [0,0,0,0,0,0,4,0,5],
          [0,0,0,7,0,2,0,0,0],
          [4,0,9,0,0,0,0,0,0],
          [2,0,0,0,1,8,0,3,0],
          [0,0,4,0,0,0,6,0,8],
          [0,0,0,4,6,0,0,0,9]
         ]
  for i in range(9):
    for j in range(9):
      trans_sudoku[i][j] = sudoku[j][i]
  return(trans_sudoku)    
    
    # Our Cost function for S
def sudoku_objective(sudoku):
  penalty = 0
  for i in range(9):
    penalty += 9-len(set(sudoku[i]))
  trans_sudoku = transpose(sudoku)
  for i in range(9):
    penalty += 9 - len(set(trans_sudoku[i]))
  for i in range(3):
    for j in range(3):
      values = set()
      for a in range(3):
        for b in range(3):
          values.add(sudoku[3*i+a][3*j+b])
      penalty += 9 - len(values)
  return(penalty)
```

```python
import pandas as pd
def make_pop_df(pop):
  Population = pd.DataFrame({'chromosome' : pop})
  Population['fitness_value'] = Population.chromosome.apply(lambda x : -sudoku_objective(x))

  min_fitness = Population.fitness_value.min()
  Population['shifted_fitness'] = Population.fitness_value - min_fitness + 0.01
  
  total_fitness = Population.shifted_fitness.sum()
  Population['Probability'] = Population.shifted_fitness / total_fitness
  Population['Cumulative_Probability']  = Population.Probability.cumsum() 
  
  return(Population) 
```

```python
def list_union(li):
    un = set()
    for x in li: un = un.union(x)
    return un
```
